###### Testons ça avec un test Chi2! On prend H0 : Les deux variables sont indépendantes. avec un niveau de confiance de 95%

# TEST CHI-SQUARE

```python
chi2_ratings = np.around(ratings,1)
chi2_bw_entropies = np.around(bw_entropies,2)
chi2_color_entropies = np.around(color_entropies,2)
chi2_rapports = np.around(rapports,2)
```

```python
from scipy.stats import chi2_contingency
from scipy.stats import chi2

test1 = np.stack((chi2_ratings, chi2_bw_entropies),axis = 1)
data_contingency1 = pd.DataFrame(data = test1)
# Contingency table.
contingency1=  pd.crosstab(data_contingency1[0], data_contingency1[1])

chi2_f, p, dof, _ = chi2_contingency(contingency1)
print("p is ", p)
print('Degree of Freedom dof=%d' % dof)
print('Chi2 = %d' % chi2_f)
confiance = 0.95
critical = chi2.ppf(confiance, dof)
print('confiance=%.2f, critical=%.3f, Chi2=%.3f' % (confiance, critical, chi2_f))
if abs(chi2_f) >= critical:
    print('Les variables sont dépendantes, on rejette H0')
else:
    print('Les variables sont indépendantes, on ne rejette pas H0')
```

###### D'après le test Chi2 sur les données, il n'y a pas une indépendance entre les variables notes et Entropy BW. 

```python
test2 = np.stack((chi2_ratings, chi2_color_entropies),axis = 1)
data_contingency2 = pd.DataFrame(data = test2)
# Contingency table.


contingency2 = pd.crosstab(data_contingency2[0], data_contingency2[1])
chi2_f, p, dof, _ = chi2_contingency(contingency2)
print("p is ", p)
print('Degree of Freedom dof=%d' % dof)
print('Chi2 = %d' % chi2_f)
confiance = 0.95
critical = chi2.ppf(confiance, dof)
print('confiance=%.2f, critical=%.3f, Chi2=%.3f' % (confiance, critical, chi2_f))
if abs(chi2_f) >= critical:
    print('Les variables sont dépendantes, on rejette H0')
else:
    print('Les variables sont indépendantes, on ne rejette pas H0')
```

# Images : Proba(Beauté,Entropie)

Vu qu'on a déjà le tableau de contingences des notes de beauté et entropies, on peut, à partir de ce tableau calculer les probas P(B,E) en normalisant sur la somme du tableau

```python
#Ici c'est pour l'image P(Beauté, Entropie BW)
contingency1 = np.array(contingency1)
contingency1 = np.where(contingency1>0,contingency1,0)#rendre le background noir
contingency1 = (contingency1*255)/contingency1.sum()#normaliser
contingency1 = np.array(list(list(x) for x in zip(*contingency1))[::-1])
imshow(contingency1, 'Image représentant la distribution (B,E_BW), proportions réelles')
#pour plus de lisibilité on étire l'image
img = np.repeat(contingency1, 10, axis = 1) #discrétiser les notes /1000
img = np.where(img>0,img,0)
imshow(img,'Image représentant la distribution (B,E_BW) étirée')
img.shape
```

```python
#Ici c'est pour l'image P(Beauté, Entropie C)
contingency2 = np.rot90(np.array(contingency2),1)
contingency2 = np.where(contingency2>0,contingency2,0)#rendre le background noir
contingency2 = (contingency2*255)/contingency2.sum()#normaliser
contingency2 = np.array(list(list(x) for x in zip(*contingency2))[::-1])
imshow(contingency2, 'Image représentant la distribution (B,E_C), proportions réelles')
img2 = np.repeat(contingency2, 20, axis = 1)
img2 = np.where(img2>0,img2,0)
imshow(img2,'Image représentant la distribution B,E_C étirée')
img2.shape
```

```python
#Ici c'est pour l'image P(Beauté, Rapport E/E_BW)
contingency3 = np.array(contingency3)
contingency3 = np.where(contingency3>0,contingency3,0) #rendre le background noir
contingency3 = (contingency3*255)/contingency3.sum() #normaliser
contingency3 = np.array(list(list(x) for x in zip(*contingency3))[::-1])
imshow(contingency3, 'Image représentant la distribution (B,Rapport E/E_BW), proportions réelles')

#Même chose, pour plus de lisibilité on étire l'image
img3 = np.repeat(contingency3, 3, axis = 1)
img3 = np.where(img3>0,img3,0)
imshow(img3,'Image représentant la distribution (B,Rapport E/E_BW) étirée')
```
